```python
import random
import time
import numpy as np

def self_avoiding_walk(steps, step_length):
    position = [0, 0]
    walk = [tuple(position)]
    directions = [(step_length, 0), (-step_length, 0), (0, step_length), (0, -step_length)]

    for _ in range(steps - 1):
        valid_moves = [d for d in directions if tuple(np.add(position, d)) not in walk]
        if not valid_moves:
            break  # No valid moves, terminate the walk
        move = random.choice(valid_moves)
        position = list(np.add(position, move))
        walk.append(tuple(position))
    
    return walk

def simulate_saw(samples, steps, step_length):
    distances = []
    times = []
    for _ in range(samples):
        start_time = time.time()
        walk = self_avoiding_walk(steps, step_length)
        end_time = time.time()
        
        if len(walk) > 1:  # Ensure the walk has more than the starting point
            distances.append(np.linalg.norm(np.subtract(walk[-1], walk[0])))
            times.append(end_time - start_time)
    
    avg_distance = np.mean(distances) if distances else 0
    distance_std = np.std(distances) if distances else 0
    avg_time = np.mean(times) if times else 0

    return avg_distance, distance_std, avg_time

# Parameters
step_lengths = [1, 2, 5]
samples = 100
steps = 100

# Analysis
for step_length in step_lengths:
    avg_distance, distance_std, avg_time = simulate_saw(samples, steps, step_length)
    print(f"Step length: {step_length}, Avg. Distance: {avg_distance}, Std. Dev: {distance_std}, Avg. Time: {avg_time}s")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random

def self_avoiding_walk(max_steps):
    moves = [(1, 0), (-1, 0), (0, 1), (0, -1)]
    walk = [(0, 0)]
    for _ in range(max_steps):
        possible_moves = [move for move in moves if (walk[-1][0] + move[0], walk[-1][1] + move[1]) not in walk]
        if not possible_moves:
            break  # No moves possible, walk ends
        next_move = random.choice(possible_moves)
        next_position = (walk[-1][0] + next_move[0], walk[-1][1] + next_move[1])
        walk.append(next_position)
    return len(walk), walk  # Return walk length and path

# Parameters
max_steps = 400
samples = 1000
walk_lengths = []

# Perform simulations
for _ in range(samples):
    length, _ = self_avoiding_walk(max_steps)
    walk_lengths.append(length)

# Calculate walk length distribution
length_counts = np.bincount(walk_lengths)
steps = np.arange(len(length_counts))

# Calculate entropy S based on the number of walks reaching each length
W_N = length_counts[length_counts > 0]  # Successful attempts for each N
entropy_S = np.log(W_N)

# Plotting
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.bar(steps, length_counts, color='skyblue')
plt.xlabel('Number of Steps N')
plt.ylabel('Number of Self-Avoiding Walks')
plt.title('Distribution of Walk Lengths')

plt.subplot(1, 2, 2)
plt.plot(steps[length_counts>0], entropy_S, marker='o', linestyle='-', color='orange')
plt.xlabel('Number of Steps N')
plt.ylabel('Entropy S - S0')
plt.title('Entropy Change with Step Number')

plt.tight_layout()
plt.show()

print("Walk lengths where walks terminated:", np.nonzero(length_counts)[0])
```

```python
import numpy as np
import random
import time
import matplotlib.pyplot as plt

def self_avoiding_walk(steps):
    directions = [(1, 0), (-1, 0), (0, 1), (0, -1)]
    walk = [(0, 0)]
    for _ in range(steps):
        last = walk[-1]
        possible_moves = [(last[0] + dx, last[1] + dy) for dx, dy in directions if (last[0] + dx, last[1] + dy) not in walk]
        if not possible_moves:
            break
        walk.append(random.choice(possible_moves))
    return walk

def simulate_walks_for_steps(step_lengths, samples):
    avg_dx = []
    avg_dy = []
    fluc_x = []
    fluc_y = []
    execution_times = []
    for step_length in step_lengths:
        distances_x, distances_y = [], []
        start_time = time.time()
        for _ in range(samples):
            walk = self_avoiding_walk(step_length)
            distances_x.append(abs(walk[-1][0] - walk[0][0]))
            distances_y.append(abs(walk[-1][1] - walk[0][1]))
        execution_time = time.time() - start_time
        avg_dx.append(np.mean(distances_x))
        avg_dy.append(np.mean(distances_y))
        fluc_x.append(np.std(distances_x))
        fluc_y.append(np.std(distances_y))
        execution_times.append(execution_time / samples)
    return avg_dx, avg_dy, fluc_x, fluc_y, execution_times

# Parameters
step_lengths = np.linspace(1, 50, 40).astype(int)
samples = 100

avg_dx, avg_dy, fluc_x, fluc_y, execution_times = simulate_walks_for_steps(step_lengths, samples)

# Plotting
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Average distances
axs[0].plot(step_lengths, avg_dx, label='Avg Dx')
axs[0].plot(step_lengths, avg_dy, label='Avg Dy')
axs[0].set_xlabel('Step Length')
axs[0].set_ylabel('Average End-to-End Distance')
axs[0].legend()
axs[0].set_title('Average End-to-End Distance vs. Step Length')

# Fluctuations
axs[1].plot(step_lengths, fluc_x, label='Fluc Dx')
axs[1].plot(step_lengths, fluc_y, label='Fluc Dy')
axs[1].set_xlabel('Step Length')
axs[1].set_ylabel('Fluctuation in End-to-End Distance')
axs[1].legend()
axs[1].set_title('Fluctuation in End-to-End Distance vs. Step Length')

# Execution times
axs[2].plot(step_lengths, execution_times, label='Execution Time')
axs[2].set_xlabel('Step Length')
axs[2].set_ylabel('Execution Time (s)')
axs[2].legend()
axs[2].set_title('Average Execution Time vs. Step Length')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt

# Function for simple sampling SAW
def self_avoiding_walk_simple(steps):
    walk = [(0, 0)]
    for _ in range(1, steps):
        current_x, current_y = walk[-1]
        moves = [
            (current_x, current_y + 1),  # Up
            (current_x, current_y - 1),  # Down
            (current_x - 1, current_y),  # Left
            (current_x + 1, current_y)   # Right
        ]
        moves = [move for move in moves if move not in walk]
        if not moves:
            break
        next_move = random.choice(moves)
        walk.append(next_move)
    return walk

# Function for biased sampling SAW
def self_avoiding_walk_biased_weighted(steps, d=2):
    walk = [(0, 0)]
    weight = 1.0
    for _ in range(1, steps):
        current_x, current_y = walk[-1]
        moves = [
            (current_x, current_y + 1),
            (current_x, current_y - 1),
            (current_x - 1, current_y),
            (current_x + 1, current_y)
        ]
        moves = [move for move in moves if move not in walk]
        l_i = len(moves)
        if l_i == 0:
            return walk, 0
        weight *= l_i / (2 * d - 1)
        next_move = random.choice(moves)
        walk.append(next_move)
    return walk, weight

# Set the number of steps for the walk
steps = 200
runs = 300
# Generate walks
simple_walk = self_avoiding_walk_simple(steps)
biased_walk, weight = self_avoiding_walk_biased_weighted(steps)

# Print important results
#print(f"Simple walk: {simple_walk}")
#print(f"Biased walk: {biased_walk}")
print(f"Weight of the biased walk: {weight}")

# Plotting
plt.figure(figsize=(10, 5))

# Plot simple sampling walk
x, y = zip(*simple_walk)
plt.plot(x, y, label='Simple Sampling', marker='o')

# Plot biased sampling walk
x, y = zip(*biased_walk)
plt.plot(x, y, label='Biased Sampling', marker='o')

plt.title('Self-Avoiding Walk: Simple vs. Biased Sampling')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.axis('equal')  # Ensure equal scaling for both axes
plt.show()
```

```python
def end_to_end_distance(walk):
    """Calculate the end-to-end distance of a walk."""
    start = walk[0]
    end = walk[-1]
    distance = np.sqrt((start[0] - end[0])**2 + (start[1] - end[1])**2)
    return distance

def calculate_average_distance(walk_function, steps, runs, *args):
    """Calculate the average end-to-end distance for a given walk function."""
    total_distance = 0.0
    valid_runs = 0
    for _ in range(runs):
        walk = walk_function(steps, *args)
        # For biased walks, walk is a tuple (walk, weight)
        if isinstance(walk, tuple):
            walk, weight = walk
            if weight > 0:  # Consider only valid walks
                total_distance += end_to_end_distance(walk) * weight
                valid_runs += weight
        else:
            total_distance += end_to_end_distance(walk)
            valid_runs += 1
    return total_distance / valid_runs if valid_runs else 0

# Set the number of steps and runs for the walks
steps = 200
runs = 500

# Calculate average end-to-end distances
simple_avg_distance = calculate_average_distance(self_avoiding_walk_simple, steps, runs)
biased_avg_distance = calculate_average_distance(self_avoiding_walk_biased_weighted, steps, runs, 2)

# Print important results
print(f"Average end-to-end distance for simple sampling: {simple_avg_distance}")
print(f"Average end-to-end distance for biased sampling: {biased_avg_distance}")

# Plotting - reuse the last walk from each method for illustration
plt.figure(figsize=(10, 5))

# Plot simple sampling walk
x, y = zip(*simple_walk)
plt.plot(x, y, label=f'Simple Sampling (Avg. Dist. {simple_avg_distance:.2f})', marker='o')

# Plot biased sampling walk
x, y = zip(*biased_walk)
plt.plot(x, y, label=f'Biased Sampling (Avg. Dist. {biased_avg_distance:.2f})', marker='o')

plt.title('Self-Avoiding Walk: Simple vs. Biased Sampling')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.axis('equal')  # Ensure equal scaling for both axes
plt.show()
```
